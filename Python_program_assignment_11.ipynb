```python
def find_k(k, str):
    string = []
    text = str.split(" ")
    for x in text:
        if len(x) > k:
            string.append(x)
    return string
k = int(input("Enter a number to compare :- "))
str = input("Enter your character :- ")
print(find_k(k, str))
```

```python
def remove(string, i):
    a = string[: i]
    b = string[i + 1:]
    return a + b
string = input("Enter your character :- ")
i = int(input("Enter your i-th number :- "))
print(remove(string, i))
```

```python
def Uncommon(A, B):
    count = {}
    for word in A.split():
        count[word] = count.get(word, 0) + 1
    for word in B.split():
        count[word] = count.get(word, 0) + 1
    return [word for word in count if count[word] == 1]
A = input("Enter your first sentance :- ")
B = input("Enter your second sentance :- ")
print(Uncommon(A, B))
```

```python
from os import sendfile

def find_dup(sen):
    x=[]
    for i in sen:
        if i not in x and sen.count(i)>1:
            x.append(i)
    print(" ".join(x)," -> is Duplicated")
sen = input("Enter your sentance :- ")
find_dup(sen)
```

```python
import re
def find(string):
    special_char=re.compile('[@_!$%^&*()<>?/\|}{~:]#')
    if special_char.search(string) == None:
        return "Yes given sentence have special character "
    else:
        return "NO given sentence have not any special character"
s=input("Enter your sentance :- ")
print(s)
print(find(s))
```
