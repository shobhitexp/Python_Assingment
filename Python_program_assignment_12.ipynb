```python
def extract_unique_values(dictionary):
    unique_values = set()
    for values in dictionary.values():
        if isinstance(values, list) or isinstance(values, set):
            unique_values.update(values)
        else:
            unique_values.add(values)
    return unique_values
example_dict = {
    'k1': [1, 2, 3],
    'k2': [2, 3, 4],
    'k3': 5,
    'k4': 6
}
unique_values = extract_unique_values(example_dict)
print("Unique values:", unique_values)
```

```python
def sum_of_items_in_dictionary(dictionary):
    ts = 0
    for value in dictionary.values():
        ts += value
    return ts
example_dict = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
}
ts = sum_of_items_in_dictionary(example_dict)
print("Sum of all items in the dictionary:", ts)
```

```python
def merge_two_dicts(dict1, dict2):
    return {**dict1, **dict2}
dict1 = {'a': 1, 'b': 2,'e':5}
dict2 = {'c': 3, 'd': 4}
merged_dict = merge_two_dicts(dict1, dict2)
print("Merged dictionary:", merged_dict)
```

```python
def convert(key_value_list):
    flat_dict = {key: value for key, value in key_value_list}
    return flat_dict
key_value_list = [('a', 1), ('b', 2), ('c', 3)]
flat_dict = convert(key_value_list)
print("Flat dictionary :- ", flat_dict)
```

```python
import collections
def insertion_at_beginning(dict, key, value):
  new_dict = collections.OrderedDict()
  new_dict.update({key: value})
  new_dict.update(dict)
  return new_dict
if __name__ == "__main__":
  dict = collections.OrderedDict([("a", 1), ("b", 2), ("c", 3)])
  new_dict = insertion_at_beginning(dict, "d", 4)
  print(new_dict)
```

```python
from collections import OrderedDict
def checkOrder(input, pattern):
    dict = OrderedDict.fromkeys(input)
    ptrlen = 0
    for key,value in dict.items():
        if (key == pattern[ptrlen]):
            ptrlen = ptrlen + 1
        if (ptrlen == (len(pattern))):
            return 'true'
    return 'false'
if __name__ == "__main__":
    input = 'engineers rock'
    pattern = 'er'
    print (checkOrder(input,pattern))
```
