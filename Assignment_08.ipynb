##Answer 3 :
* inputInt() : Accepts an integer value. This also takes additional parameters 'min', 'max', 'greaterThan' and 'lessThan' for bounds. Returns an int.
* inputFloat() function:

Use inputFloat() when you expect the user to provide a floating-point number input.
It validates the user's input to ensure it can be converted to a float. If the input is not a valid floating-point number, an error message is displayed, and the user is prompted again until a valid float is entered.
It returns the validated float value.

```python
import pyinputplus as pypi
num = pypi.inputInt("Enter an integer values :- ")
print("Entered integer:", num)


num = pypi.inputFloat("Enter a float values :- ")
print("Entered float:", num)
```

##Answer 5 :-
We can also use regular expressions to specify whether an input is allowed or not. The allowRegexes and blockRegexes
keyword arguments take a list of regular expression strings to determine what the PyInputPlus function will accept or reject as valid input.

```python
response = pplus.inputNum(allowRegexes=[r'(I|V|X|L|C|D|M)+', r'zero']) # it allowes roman letters as numbers too
```

```python
response = pplus.inputNum(blockRegexes=[r'[02468]$'])# blocks the even numbers
```
